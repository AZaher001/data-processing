```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

```python
df = pd.read_csv('titanic_full.csv')
df.head()
```

```python
def title(data) :
    if 'Mr.' in data : 
        return 'Mr' 
    elif 'Mrs.' in data : 
        return 'Mrs' 
    elif 'Miss.' in data : 
        return 'Miss' 
    else:
        return 'Nothing'
```

```python
df['title']= df['Name'].apply(title)
pd.DataFrame(data = df['title'].value_counts())
```

```python
sns.countplot(data = df , x='title')
```
