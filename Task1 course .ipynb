```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

```python
df = pd.read_csv('titanic_full.csv')
df.head()
```

```python
def title(data) :
    x = data.split()
    return x[1]
```

```python
df['title']= df['Name'].apply(title)
pd.DataFrame(data = df['title'].value_counts())
```

```python
sns.countplot(data = df , y='title')
```
